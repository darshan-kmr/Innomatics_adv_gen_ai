```python
def engagement(likes):
    t=0
    for l in likes:
        t+=l
    status="viral post" if t>=1000 else "normal engagement"
    print("total likes:",t)
    print("post status:",status)
engagement([200,300,550])
```

```python
def stock(m):
    status="low stock alert" if m<10 else "stock sufficient"
    print("medicine stock:",m)
    print("status:",status)
stock(6)
```

```python
def rainfall(data,req):
    t=0
    for r in data:
        t+=r
    avg=t/len(data)
    status="adequate rainfall" if avg>=req else "inadequate rainfall"
    print("average rainfall:",avg)
    print("rainfall status:",status)
rainfall([60,80,76],70)
```

```python
def check(users):
    status="yes" if len(users)!=len(set(users)) else "no"
    print("duplicate accounts found:",status)
check(["user1","user2","user1"])
```

```python
def eligibility(age):
    status="eligible" if age>=18 else "not eligible"
    print("patient age:",age)
    print("eligibility status:",status)
eligibility(21)
```

```python
def premium(prices):
    p=[]
    for x in prices:
        if x>2000:
            p.append(x)
    print("premium crops:",p)
premium([1500,2500,3200])
```

```python
def health(errors):
    if errors==0:
        status="healthy"
    elif errors<=5:
        status="minor issues"
    else:
        status="critical issues"
    print("error count:",errors)
    print("system status:",status)
health(7)
```

```python
def transaction(amt):
    status="approved" if amt<=50000 else "rejected"
    print("transaction amount:",amt)
    print("transaction status:",status)
transaction(60000)
```

```python
def attendance(a):
    c=0
    for x in a:
        if x=="P":
            c+=1
    per=(c/len(a))*100
    status="eligible" if per>=75 else "not eligible"
    print("attendance percentage:",per)
    print("exam eligibility:",status)
attendance(["P","P","A","P","P"])
```
