```python
sen=input("Enter sentence:")
words=sen.split()   
uni_words = set(words)
print("Unique words count:",len(uni_words))
print("Unique words:",uni_words)
```

```python
employees={"Ravi": 75000,"Anita": 68000,"Kiran": 72000}
high_name=max(employees, key=employees.get)
high_sal=employees[high_name]
print("Highest Salary:",high_name,"-",high_sal)
```

```python
prices=[450,1200,899,1500,300]
count=sum(1 for price in prices if price>1000)
print("Products above 1000:",count)
```
